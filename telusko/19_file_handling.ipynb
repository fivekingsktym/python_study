#### File Handling in python

```python
# Reading a file
f = open('19_file_handling.txt', 'r') # file is opening for read mode

print(type(f))
print(f, '\n')


# print(f.read()) # reading a the full data in the file
print(f.readline()) # reading the first line in the file
print(f.readline(9)) # reading the second line with a string len 9, why the line is second, the pointer will increment to new line
```

```python
# Writing to a new file

f = open('19.2_file_write_example.txt', 'w')
f.write('hello world')
f.close()
```

```python
# Append a string to an existing file

f = open('19.2_file_write_example.txt', 'a')
f.write('\nsecond line')
f.close()
```

```python
# Reading a file with binary format and copying and saving to another file
f1 = open('19.3_catimage.jpg', 'rb') # reading file
# print(f.read())

f2 = open('19.4_catimage_binary.jpg', 'wb') # writing file

for i in f1:
    f2.write(i) # copying fileq and saving the file2
    print(i)
```
